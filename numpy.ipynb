###Ajustar uma reta para entender a taxa de crescimento do preço das maçãs

y=ax+b

a = coeficiente angular - Responsável por dar o ângulo da reta

b = coeficiente linear - Onde a reta corta o eixo y

OBS: y é o eixo dos preços das maçãs e x é o eixo dos meses

```python
x = datas
```

```python
y = 2*x + 80
```

```python
plt.plot(datas, Moscow)
plt.plot(x, y)
```

```python
np.sqrt(np.sum(np.power(Moscow - y, 2))) #power - eleva valores a uma potência | sum - somar os valores | sqrt - retorna a raiz quadrada dos valores
```

###Calculando o coeficiente angular:

```python
Y = Moscow
X = datas
n = np.size(Moscow)
```

```python
a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2) - np.sum(X)**2)
```

###Calculando o coeficiente linear;

```python
b = np.mean(Y) - a*np.mean(X)
```

```python
y = a*X+b
```

```python
np.linalg.norm(Moscow-y)
```

```python
plt.plot(datas, Moscow)
plt.plot(x, y)
plt.plot(41.5, 41.5*a+b, '*r')
plt.plot(100,100*a+b, '*r')
```
